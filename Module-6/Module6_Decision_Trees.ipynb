# Module 6: Decision Analysis & Decision Trees


In this module, we explore the structured approach of **decision-making** using **decision trees**. Decision trees are powerful visual tools that help simplify complex choices and outcomes. They break down problems into manageable components, consisting of:

- **Decision nodes** (squares) for choices,
- **Chance nodes** (circles) for uncertain outcomes,
- **End nodes** (triangles) for final results.

By mapping decisions in a tree structure, we can systematically evaluate different paths of action.



## Overview of Decision Analysis

Decision analysis is a structured approach for making choices in complex and uncertain situations. It is used across various fields like business, healthcare, and environmental planning. The process typically involves:

1. **Defining the issue**  
2. **Identifying alternatives**  
3. **Gathering relevant information**  
4. **Estimating probabilities**  
5. **Evaluating consequences**  
6. **Selecting the best action** based on preferences and risk tolerance

This process helps reduce biases and improve decision quality.



## Fundamentals of Decision Trees


To construct a decision tree:
- Start from the root decision node
- Branch out all possible alternatives
- For each chance event, assign probabilities and outcomes
- Roll back the tree using **expected values**



## Example: Investment Decision Tree

An investor must decide whether to invest in **stocks** or **bonds**.

### Probabilities and Payoffs:

- **Stocks**  
  - 60% chance of earning **$10,000**  
  - 40% chance of losing **$5,000**  
- **Bonds**  
  - 30% chance of earning **$3,000**  
  - 70% chance of earning **$2,000**

### Expected Values:
- **Stocks**:  
  $$EV = 0.6 \times 10,000 + 0.4 \times (-5,000) = 6,000 - 2,000 = 4,000$$

- **Bonds**:  
  $$EV = 0.3 \times 3,000 + 0.7 \times 2,000 = 900 + 1,400 = 2,300$$

**Decision:** Invest in **stocks** for a higher expected value.


## Probability in Decision Analysis

Probability quantifies the likelihood of events and is critical to modeling uncertainty.

### Definitions of Probability

- **Classical**:  
  $$
  P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}
  $$

- **Frequentist**:  
  $$
  P(A) = \lim_{n \to \infty} \frac{\text{Number of times A occurs}}{\text{Total trials}}
  $$

- **Subjective**:  
  Based on expert belief or intuition.

- **Axiomatic**: 
It's based on 3 main axioms:

  - $ P(\Omega) = 1         ((\Omega) $  is called Omega - which means that the propbability of entire sample space quals to one) 
  - $ P(A \cup B) = P(A) + P(B) $  (if A, B are mutually exclusive)


### Variance & Standard Deviation in Decision Analysis

####  Variance

Variance measures how far a set of values is spread out from their mean. It is calculated using the formula:

$$
\sigma^2 = \frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2
$$

Where:

- $ \sigma^2 $: Population variance  
- $ x_i $: Each individual value  
- $ \mu $: Mean of the dataset  
- $ N $: Total number of values  

####  Standard Deviation

Standard deviation is the square root of the variance:

$$
\sigma = \sqrt{\sigma^2}
$$

It represents how much the values in a dataset deviate from the mean. Standard deviation is useful for understanding the spread or consistency of data. It's in the same unit of data!



---

In the example below, we simulate student grades to calculate both variance and standard deviation.


```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated student grades: some low, some high, one perfect
grades = np.array([0,1,2,3,4,4,4, 5, 6, 7, 70, 75, 80, 95,100])

# Calculate statistics
mean = np.mean(grades)
variance = np.var(grades)  # Population variance
std_dev = np.std(grades)   # Population standard deviation

# Print results
print(f"Grades: {grades}")
print(f"Mean: {mean:.2f}")
print(f"Variance (): {variance:.2f}")
print(f"Standard Deviation (): {std_dev:.2f}")

# Visualize the grade distribution
plt.figure(figsize=(10, 5))
plt.bar(range(len(grades)), grades, color='skyblue', edgecolor='black')
plt.axhline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.title('Simulated Student Grades')
plt.xlabel('Student Index')
plt.ylabel('Grade')
plt.xticks(range(len(grades)))
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Expected Value & Decision Criteria in Decision Analysis

### Calculating Expected Monetary Value (EMV)
Expected Monetary Value (EMV) is a key concept in decision analysis, used to quantify the average outcome of an uncertain event.

EMV comes from the concept of Expected Value in Probability Theory.  
It is calculated by multiplying each possible outcome by its probability of occurrence and then summing these products:

$$
EMV = \sum_{i=1}^n [ \text{Probability of outcome}_i \times \text{Monetary value of outcome}_i ]
$$

---

### Risk Attitudes

**Risk attitude** refers to a decision-maker's approach to uncertain outcomes:

- **Risk-Neutral**  
  - Makes decisions solely based on EMV.  
  - Indifferent between a certain outcome and an uncertain one with the same EMV.  
  - *Example:* Equally happy with a guaranteed \$500 or a 50% chance of \$1000.

- **Risk-Averse**  
  - Prefers a more certain outcome over an uncertain one with the same or slightly higher EMV.  
  - Willing to accept a lower EMV for reduced risk.  
  - *Example:* Prefers a guaranteed \$450 over a 50% chance of \$1000.

- **Risk-Seeking**  
  - Prefers an uncertain outcome over a certain one with the same or slightly lower EMV.  
  - Willing to accept higher risk for the chance of a better outcome.  
  - *Example:* Prefers a 50% chance of \$1000 over a guaranteed \$550.

---

### Other Decision Criteria

1. **Minimax**  Minimize the maximum possible loss.  
   - Pessimistic approach, preparing for the worst-case scenario.  
   - Formula: Choose the option with the smallest maximum loss.  
   - Useful in competitive situations or when avoiding worst outcomes is crucial.

2. **Maximin**  Maximize the minimum possible gain.  
   - Conservative approach, ensuring the best worst-case scenario.  
   - Formula: Choose the option with the largest minimum gain.  
   - Useful when ensuring a minimum level of success is critical.

3. **Regret Minimization**  
   - Minimizes the feeling of *"what could have been"* after the decision.  
   - Steps:  
     1. Calculate the regret for each outcome (difference between the best possible outcome and the actual outcome).  
     2. Determine the maximum regret for each decision alternative.  
     3. Choose the alternative with the smallest maximum regret.


```python
# Example: Calculating EMV for two investment options
# Stocks: 60% chance of +$10,000, 40% chance of -$5,000
# Bonds:  30% chance of +$3,000, 70% chance of +$2,000

# Define outcomes and probabilities
stocks_outcomes = [10000, -5000]
stocks_probs = [0.6, 0.4]

bonds_outcomes = [3000, 2000]
bonds_probs = [0.3, 0.7]

# Calculate EMV
emv_stocks = sum(p * outcome for p, outcome in zip(stocks_probs, stocks_outcomes))
emv_bonds = sum(p * outcome for p, outcome in zip(bonds_probs, bonds_outcomes))

print(f"EMV (Stocks): ${emv_stocks:,.2f}")
print(f"EMV (Bonds): ${emv_bonds:,.2f}")

# Decision recommendation
if emv_stocks > emv_bonds:
    print("Recommendation: Invest in Stocks (higher EMV).")
elif emv_bonds > emv_stocks:
    print("Recommendation: Invest in Bonds (higher EMV).")
else:
    print("Both options have equal EMV; choose based on risk preference.")

# Explanation of the logic:
# 1. The code calculates the Expected Monetary Value (EMV) for two investment options: stocks and bonds.
# 2. EMV is calculated by multiplying each outcome by its probability and summing the results.
#    - For stocks: EMV = (0.6 * 10000) + (0.4 * -5000)
#    - For bonds: EMV = (0.3 * 3000) + (0.7 * 2000)
# 3. The EMV represents the average monetary value of each investment option based on the given probabilities.
# 4. The code compares the EMVs of stocks and bonds to recommend the option with the higher EMV.
# 5. If both options have the same EMV, the decision is left to the investor's
```


## Example of Regret Minimization
 





### 1. What is Regret?
Regret is the **opportunity loss**  the amount you "miss out on" compared to the best possible payoff for a given situation.  

For each market condition:
- Find the **best payoff** among all options.
- Subtract each options payoff from this best payoff.  
- This difference is the regret for that option in that market condition.

---

### 2. Constructing the Regret Table
**Bull Market**:  
- Best payoff = $10,000 (Investment A)  
- Regret for A = $10,000  $10,000 = $0  
- Regret for B = $10,000  $8,000 = $2,000  

**Bear Market**:  
- Best payoff = $5,000 (Investment B)  
- Regret for A = $5,000  $2,000 = $3,000  
- Regret for B = $5,000  $5,000 = $0  

**Regret Table:**

| Market Condition | Investment A | Investment B |
|------------------|--------------|--------------|
| Bull Market      | $0           | $2,000       |
| Bear Market      | $3,000       | $0           |

---

### 3. Maximum Regret Rule
The **Regret Minimization** (Minimax Regret) approach says:  

Choose the option with the **smallest maximum regret**.

- Investment A: Max regret = max($0, $3,000) = **$3,000**  
- Investment B: Max regret = max($2,000, $0) = **$2,000**
**Decision**: Since $2,000 < $3,000, choose **Investment B**.


## Value of Information in Decision Analysis

Decision analysis often considers **the value of information**how much having additional information improves decision-making outcomes.

---

### Perfect Information & Its Value

**Perfect information** refers to knowing with certainty what the outcome of a decision will be **before making it**. While rarely attainable in real-world scenarios, it serves as a theoretical benchmark.

**Definition:**
The **Value of Perfect Information (VPI)** is the difference between:
- The expected value of the best decision **with perfect information**  
- The expected value of the best decision **without perfect information**

**Formula:**
$$
VPI = E[\max(\text{outcomes with perfect info})] - \max(E[\text{outcomes without perfect info}])
$$



---

### Imperfect Information & Its Value

In practice, perfect information is rare. Instead, we often have access to **imperfect information** that improves our decision-making but does **not** guarantee the outcome.

**Definition:**
The value of imperfect information is the **increase in expected value** due to partially reliable information.

---

#### Calculation Process:
1. **Determine the reliability** of the information (e.g., accuracy of a forecast or diagnostic test).
2. **Update probabilities** using **Bayes' Theorem**:
   $$
   P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
   $$
3. **Recalculate expected values** with updated probabilities.
4. **Compare** the new expected value with the original to find the value of imperfect information.

---

#### Example:
Continuing with a stock market investment scenario:

- Forecast accuracy: **80%**
- Prior probabilities:  
  $$ P(\text{Bull}) = 0.6, \quad P(\text{Bear}) = 0.4 $$

**Bayesian updating:**
$$
P(\text{Bull}|\text{Forecast Bull}) = \frac{0.8 \cdot 0.6}{(0.8 \cdot 0.6) + (0.2 \cdot 0.4)} \approx 0.86
$$
$$
P(\text{Bear}|\text{Forecast Bear}) = \frac{0.8 \cdot 0.4}{(0.2 \cdot 0.6) + (0.8 \cdot 0.4)} \approx 0.73
$$

**Expected values:**
- If forecast = Bull:
  $$
  EV = 0.86 \cdot 10,000 + 0.14 \cdot 2,000 = 8,880
  $$
- If forecast = Bear:
  $$
  EV = 0.27 \cdot 10,000 + 0.73 \cdot 5,000 = 6,350
  $$

**Overall expected value with imperfect info:**
$$
0.6 \cdot 8,880 + 0.4 \cdot 6,350 = 7,868
$$

**Value of imperfect information:**
$$
V_{\text{imperfect}} = 7,868 - 6,800 = 1,068
$$
