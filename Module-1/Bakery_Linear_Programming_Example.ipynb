# Linear Programming: The Bakery Optimization Problem

Suppose you run a bakery and want to maximize your daily profit by deciding how many **cakes** (x) and **cookies** (y) to bake.

- Each cake earns **$5** profit.
- Each cookie earns **$2** profit.
- Each cake takes **2 hours** to bake.
- Each cookie takes **0.5 hours** to bake.
- You have **8 hours** of baking time available.

## Objective Function
$$
\text{Maximize}\quad z = 5x + 2y
$$
where:
- $x$ = number of cakes
- $y$ = number of cookies

## Constraints
$$
\begin{align*}
2x + 0.5y &\leq 8 \quad\text{(time constraint)} \\
x &\geq 0 \quad\text{(no negative cakes)} \\
y &\geq 0 \quad\text{(no negative cookies)}
\end{align*}


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 200)
y = 16 - 4 * x  # From constraint: 2x + 0.5y = 8

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$2x + 0.5y = 8$')
plt.fill_between(x, 0, np.clip(y, 0, 16), color='lightgreen', alpha=0.5, label='Feasible Region')

# Mark the vertices
plt.scatter([0, 4, 0], [0, 0, 16], color='red', zorder=5)
plt.text(0, 0, "(0, 0)", fontsize=10, verticalalignment='top', horizontalalignment='right')
plt.text(4, 0, "(4, 0)", fontsize=10, verticalalignment='top', horizontalalignment='left')
plt.text(0, 16, "(0, 16)", fontsize=10, verticalalignment='bottom', horizontalalignment='right')

plt.xlim(-0.5, 5)
plt.ylim(-1, 18)
plt.xlabel('Cakes (x)')
plt.ylabel('Cookies (y)')
plt.title('Feasible Region for the Bakery Problem')
plt.legend()
plt.grid(True)
plt.show()
```

## The Feasible Region
The feasible region is the triangle where all constraints are satisfied:
- Below the line $2x + 0.5y = 8$
- $x \geq 0$ and $y \geq 0$

The **vertices** (corners) of this region are:
- (0, 0)
- (4, 0)
- (0, 16)


## Evaluating the Objective Function at Each Vertex

| Vertex    | Calculation                | Profit ($z$) |
|-----------|----------------------------|--------------|
| (0, 0)    | $5*0 + 2*0$                | 0            |
| (4, 0)    | $5*4 + 2*0$                | 20           |
| (0, 16)   | $5*0 + 2*16$               | 32           |

So, the **maximum profit** is **$32**, by baking **0 cakes** and **16 cookies**.

## Key Learning: The Beautiful Rule of Linear Programming

> **If the objective function and all constraints are linear,  
> the maximum (or minimum) value of the objective function  
> will always occur at one of the **vertices (corners)** of the feasible region.**

This is why we check only the corner points!

## Summary of Learnings

- **Vertex / Vertices:** These are the corner points of your feasible region (solution area).
- **Feasible Region:** The set of all points (solutions) that satisfy all constraints.
- **Objective Function:** The quantity to maximize or minimize (here: profit).
- **Constraints:** The mathematical representation of real-world limits.

### Steps to Solve Any Linear Programming Problem
1. **Define variables and the objective function.**
2. **Write out the constraints.**
3. **Plot the feasible region and find its vertices.**
4. **Evaluate the objective function at each vertex.**
5. **Choose the vertex with the best value for your goal.**

This method works for any linear programming problem!
